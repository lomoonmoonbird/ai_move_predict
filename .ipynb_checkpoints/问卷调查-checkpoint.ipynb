```python
import pandas as pd
import numpy as np 

#原始数据表格清洗
pd.set_option('mode.chained_assignment', None)
#pd.set_option('precision', 0)
pd.options.display.precision=0
df = pd.read_excel("./问卷调查.xlsx", skiprows=2)
df.dropna(how='all',axis=0, inplace=True)
df.dropna(how='all',axis=1, inplace=True)

health_reg = ["装修","甲醛", "污染","超标", "有害", "气体","装修"]
transport_reg = ["停车","交通", "通勤", "油费", "km", "四环","三环", "时间太长", "距离远",
"限行","搬迁","工作制","租房", "距离","小时","停车位","充电"] 
eat_reg = ["餐补","吃饭", "饭店", "生活","设施","餐饮"]

df["搬迁对自己造成的问题1："].fillna("", inplace=True)
df.loc[df["搬迁对自己造成的问题1："].str.contains('|'.join(health_reg)) 
| df["搬迁对自己造成的问题2："].str.contains('|'.join(health_reg)) 
| df["搬迁对自己造成的问题3："].str.contains('|'.join(health_reg)), "健康"] = 1

df["搬迁对自己造成的问题2："].fillna("", inplace=True)
df.loc[df["搬迁对自己造成的问题1："].str.contains('|'.join(transport_reg)) 
| df["搬迁对自己造成的问题2："].str.contains('|'.join(transport_reg)) 
| df["搬迁对自己造成的问题3："].str.contains('|'.join(transport_reg)), "通勤"] = 1

df["搬迁对自己造成的问题3："].fillna("", inplace=True)
df.loc[df["搬迁对自己造成的问题1："].str.contains('|'.join(eat_reg)) 
| df["搬迁对自己造成的问题2："].str.contains('|'.join(eat_reg)) 
| df["搬迁对自己造成的问题3："].str.contains('|'.join(eat_reg)), "吃饭"] = 1

df["对此有何意见/建议"].fillna("", inplace=True)
df.loc[df["对此有何意见/建议"].str.contains('|'.join(health_reg)) 
| df["对此有何意见/建议"].str.contains('|'.join(health_reg)) 
| df["对此有何意见/建议"].str.contains('|'.join(health_reg)), "健康"] = 1

df.loc[df["对此有何意见/建议"].str.contains('|'.join(transport_reg)) 
| df["对此有何意见/建议"].str.contains('|'.join(transport_reg)) 
| df["对此有何意见/建议"].str.contains('|'.join(transport_reg)), "通勤"] = 1

df.loc[df["对此有何意见/建议"].str.contains('|'.join(eat_reg)) 
| df["对此有何意见/建议"].str.contains('|'.join(eat_reg)) 
| df["对此有何意见/建议"].str.contains('|'.join(eat_reg)), "吃饭"] = 1

df["其他"].fillna("", inplace=True)
df.loc[df["其他"].str.contains('|'.join(health_reg)) 
| df["其他"].str.contains('|'.join(health_reg)) 
| df["其他"].str.contains('|'.join(health_reg)), "健康"] = 1

df.loc[df["其他"].str.contains('|'.join(transport_reg)) 
| df["其他"].str.contains('|'.join(transport_reg)) 
| df["其他"].str.contains('|'.join(transport_reg)), "通勤"] = 1

df.loc[df["其他"].str.contains('|'.join(eat_reg)) 
| df["其他"].str.contains('|'.join(eat_reg)) 
| df["其他"].str.contains('|'.join(eat_reg)), "吃饭"] = 1

df["其他"].fillna("", inplace=True)
df["健康"].fillna(0, inplace=True)
df["通勤"].fillna(0, inplace=True)
df["吃饭"].fillna(0, inplace=True)

agree = df["搬迁满意度（赞成/一般/不赞成）"].value_counts()
health = df["健康"].value_counts()
transport = df["通勤"].value_counts()
eat = df["吃饭"].value_counts()
health_transport = df[(df["健康"] == 1) & (df["通勤"] == 1)].value_counts()
health_eat = df[(df["健康"] == 1) & (df["吃饭"] == 1)].value_counts()
transport_eat = df[(df["通勤"] == 1) & (df["吃饭"] == 1)].value_counts()
health_transport_eat = df[(df["健康"] == 1) & (df["通勤"] == 1) & (df["吃饭"] == 1)].value_counts()

def class_transform(series):
    if series['搬迁满意度（赞成/一般/不赞成）'] == '赞成':
        agree = 1
    elif series['搬迁满意度（赞成/一般/不赞成）'] == '不赞成':
        agree = 0
    elif series['搬迁满意度（赞成/一般/不赞成）'] == '一般':
        agree = 2
    else:
        agree = 0
    return agree

df['class'] = df.apply(class_transform, axis=1)
df['理由1'] = np.where(df['搬迁对自己造成的问题1：'] !='', 1, 0)
df['理由2'] = np.where(df['搬迁对自己造成的问题2：'] !='', 1, 0)
df['理由3'] = np.where(df['搬迁对自己造成的问题3：'] !='', 1, 0)
df['理由4'] = np.where(df['对此有何意见/建议'] !='', 1, 0)
df['理由5'] = np.where(df['其他'] !='', 1, 0)
df['理由数'] = df[[ '理由1', '理由2','理由3', '理由4','理由5']].sum(axis=1)

df[15:30]
```

```python
print(agree.to_string())
print("关心健康数量:",health[1])
print("关心通勤数量:", transport[1])
print("关心吃饭数量:",eat[1])
print("关心健康和通勤数量:",len(health_transport))
print("关心健康和吃饭数量:", len(health_eat))
print("关心通勤和吃饭数量:", len(transport_eat))
print("关心健康和通勤和吃饭数量:", len(health_transport_eat))
```

```python
import  matplotlib  as plt
import numpy as np
df_copy = df.copy()

clean_table = {
    "健康":df['健康'],
    "通勤":df['通勤'],
    "吃饭":df['吃饭'],
    "理由数": df['理由数'],
    "class": df['class']
}
clean_table_df = pd.DataFrame(clean_table)
clean_table_df = clean_table_df.sample(frac=1)
clean_table_df[20:30]
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import accuracy_score
from sklearn import svm
#特征
X = clean_table_df.loc[:, clean_table_df.columns !='class']
#标签
y = clean_table_df.loc[:, clean_table_df.columns == 'class']
#分离训练和测试集 比例4:1
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=10)
print("原始训练集包含样本数量：", len(X_train))
print("原始测试集包含样本数量:", len(X_test))

# 训练 
#逻辑回归
#lm = LogisticRegression(C=10,solver='liblinear', random_state=0) #linear
#lm = LogisticRegression(multi_class = 'multinomial', solver = 'lbfgs') #multinomial
lm = linear_model.LogisticRegression(multi_class='ovr', solver='sag', max_iter=1000, random_state=42) #ovr
lm.fit(X_train, y_train.values.reshape(len(y_train), ))
y_pred_test = lm.predict(X_test)
score_reg = lm.score(X_test, y_test)
print("reg 精确率 : %f"%score_reg)
# kernel = 'rbf'
lf_rbf = svm.SVC(kernel='rbf')
lf_rbf
lf_rbf.fit(X_train, y_train.values.reshape(len(y_train), ))
score_rbf = lf_rbf.score(X_test,y_test)
print("rbf 精确率 : %f"%score_rbf)

# kernel = 'linear'
lf_rbf = svm.SVC(kernel='linear')
lf_rbf.fit(X_train, y_train.values.reshape(len(y_train), ))
score_linear = lf_rbf.score(X_test,y_test)
print("linear 精确率 : %f"%score_linear)

#kernel = 'poly'
lf_rbf = svm.SVC(kernel='poly',C=1.0, degree=3, gamma=1/4,
coef0=0.0, shrinking=True, probability=False, tol=0.001, cache_size=200,   
class_weight=None, verbose=False, max_iter=-1, decision_function_shape="ovr", 
random_state=None)
lf_rbf.fit(X_train,y_train)
score_poly = lf_rbf.score(X_test,y_test)
print("poly 精确率 : %f"%score_poly)

#预测
y_pred_train = lm.predict(X_train)
print("实际结果", np.array(y_test['class']))
print("预测结果",y_pred_test)
print('模型预测测试集准确度: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
print('模型预测训练集准确度: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
print("大鹏哥想搬吗：", {2:"一般", 1:"赞成", 0:"不赞成"}.get(lm.predict(np.array(X.loc[[103]]))[0]))
df.loc[[103]]
```
